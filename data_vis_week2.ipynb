# **Draw Multiple Data in one chart**

하나의 차트에 여러 개의 데이터를 중첩해서 그리는 것도 데이터 분석에 용이하다. 예를 들어서 리스트 x를 다음과 같이 선언하고 이를 리스트 x,y,z와 함께 그려보자.

```python
import matplotlib.pyplot as plt

x = [x for x in range(20)] #0에서 20까지의 정수를 생성해라 generate int from 0 to 20 and put in x
y = [x**2 for x in range(20)]  #0에서 20까지의 정수 x에 대해 x 제곱값을 생성해 y 리스트에 넣는다
z= [x**3 for x in range(20)]  #0에서 20까지의 정수 x에 대해 x 세제곱값을 생성해 z 리스트에 넣는다

plt.plot(x,x, label='linear')
plt.plot(x,y, label='quadratic')
plt.plot(x,z, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("My Plot")
plt.legend()  #디폴트 위치에 범례(각 차트를 설명)를 표시 label explanation for default place
plt.show()
```

# **just trying bar chart**

```python
from matplotlib import pyplot as plt

years=[1950,1960,1970,1980,1990,2000,2010]
gdp= [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.bar(range(len(years)),gdp)

plt.title("GDP per capita")
plt.ylabel("dollars")

plt.xticks(range(len(years)),years)
plt.show()
```

The range() function creates a range of integers. In this case, the years list has 7 entries, so it creates a range of integers from 0 to 6. This is the range of the horizontal axis of the bar chart.

However, the horizontal axis is not an integer from 0 to 6, but a year should be assigned to each tick. We used the xticks() function to specify the tick value to be assigned to each tick in the horizontal axis range.


# **Plot national income trends across countries in a multi-bar chart**

let's say we have 3 different countries' data and compare them

```python
years= [1965,1975, 1985, 1995, 2005, 2015]
a= [130,650,2450,11600,17790,27250]
b= [890, 5120,11500,42130,40560,38780]
c= [100,200,290,540,1760,7940]

x_range= range(len(years))
plt.bar(x_range,a,width=0.25)
plt.bar(x_range,b,width=0.25)
plt.bar(x_range,c,width=0.25)
plt.show()
```

we can't see them because they are overlapping. We are going to use numpy library's arange() to make it generate decimal numbers

```python
import numpy as np
import matplotlib.pyplot as plt

years= [1965,1975, 1985, 1995, 2005, 2015]
a= [130,650,2450,11600,17790,27250]
b= [890, 5120,11500,42130,40560,38780]
c= [100,200,290,540,1760,7940]


x_range= np.arange(len(years))    # years의 길이만큼 소수점 단위도 가능한 범위 생성. check ppt's 6th page
plt.bar(x_range+0.0,a,width=0.25)
plt.bar(x_range+0.3,b,width=0.25)
plt.bar(x_range+0.6,c,width=0.25)
plt.show()
```

# **See the distribution of your data at a glance with histograms**

  A histogram is a bar graph that divides a given set of data into bins and examines the number of frequencies in each bin.

자료의 분포를 한눈에 알아볼 수 있다.

To create a histogram, you need to divide the values in your data into equal 'bins'.

```python
import matplotlib.pyplot as plt

books=[1,6,2,3,1,2,0,2]

plt.hist(books, bins=6)

plt.xlabel("books")
plt.ylabel("frequency")
plt.show()
```

# **Multiple graphs in one picture**

use subplot() in pyplot

plt.subplots(2,2) means i am going to divide the figure into 2 by 2 grid.

One of the return values of subplots(), ax, contains the split charts.

ax[0,0] means the first one from the 2 by 2 grid. this is the indexing method

fig, ax: The function returns two objects:

fig refers to the entire figure object. It's useful for operations that affect the figure as a whole, such as saving the figure to a file.
ax is an array of axes objects, with one axes object for each subplot.

```python
import matplotlib.pyplot as plt

fig, ax= plt.subplots(2,2,figsize=(5,5))
ax[0,0].plot(range(10),'r')  #row 0, col 0
ax[1,0].plot(range(10), 'b') #row 1, col 0
ax[0,1].plot(range(10),'g')  #row 0, col 1
ax[1,1].plot(range(10), 'k') #row 1, col 1
plt.show()
```
